```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('airfoil_self_noise.dat',sep='\t',header = None)
df.columns = ['Frequency',
              'Angle of attack', 
              'Chord length', 
              'Free-stream velocity', 
              'Suction side displacement thickness',
             'Scaled sound pressure level']
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]
x
```

```python
##### train test split ######
from sklearn.model_selection import train_test_split
x_train , x_test , y_train, y_test = train_test_split(x, y ,test_size = 0.3)
```

```python
import seaborn as sns
sns.pairplot(x_train)
```

```python
import matplotlib.pyplot as plt
x_train.corr()
```

```python
for f in x_train.columns:
    sns.boxplot(x_train[f])
    plt.show()
```

```python
for i in ['Suction side displacement thickness', 'Frequency']:
    sns.scatterplot(data=df, x = i, y = 'Scaled sound pressure level')
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_train
```

```python
x_test = scaler.transform(x_test)
x_test
```

```python
from sklearn.linear_model import LinearRegression
lin_regressor  = LinearRegression()
```

```python
lin_regressor.fit(x_train , y_train)
test_predictions = lin_regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,accuracy_score,r2_score
MAE=mean_absolute_error(y_test,test_predictions)
MSE=mean_squared_error(y_test,test_predictions)
RMS=np.sqrt(MSE)

print(MAE,MSE,RMS,r2_score(y_test,test_predictions))
```

```python
from sklearn.linear_model import ElasticNet
elasticnet = ElasticNet()
```

```python
elasticnet.fit(x_train , y_train)
test_predictions = elasticnet.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,accuracy_score,r2_score
MAE=mean_absolute_error(y_test,test_predictions)
MSE=mean_squared_error(y_test,test_predictions)
RMS=np.sqrt(MSE)

print(MAE,MSE,RMS,r2_score(y_test,test_predictions))
```

```python
from sklearn.ensemble import RandomForestRegressor
rforest = RandomForestRegressor()

rforest.fit(x_train , y_train)
test_predictions = rforest.predict(x_test)

from sklearn.metrics import mean_absolute_error,mean_squared_error,accuracy_score,r2_score
MAE=mean_absolute_error(y_test,test_predictions)
MSE=mean_squared_error(y_test,test_predictions)
RMS=np.sqrt(MSE)

print(MAE,MSE,RMS,r2_score(y_test,test_predictions))

import pickle
pickle.dump(rforest,open('rforstst.pkl','wb'))

pickled_model = pickle.load(open('rforstst.pkl','rb'))
pickled_model.predict(x_test)   #### batch input
```
