## Preparing the Dataset

```python
dataset = pd.DataFrame(boston.data,columns=boston.feature_names)
dataset.head()
```

```python
dataset['Price']=boston.target
dataset.head()
```

### Exploratory Data Analysis


```python
dataset_corr = pd.DataFrame(dataset.corr()) ### correlation between the features
dataset_corr
```

```python
dataset_corr[dataset_corr>0.85] ### Checking for independent features which are higly correlated
```

```python
sns.scatterplot(dataset['TAX'], dataset['RAD'])
```

```python
dataset = dataset.drop(columns=['TAX'], axis=1) ### We choose to drop either 'RAD' or 'TAX' feature as these are highly correlated
dataset
```

### Independent dependent Features

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

### Train Test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=43)
```

### Standardizing the dataset

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

```python
import pickle
pickle.dump(scalar,open('scaling.pkl','wb'))
```

## Model Training

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,mean_absolute_error
```

```python
def model_training(model,x,y):
    model.fit(x,y)
    y_predict_train = model.predict(x)
    print(f'For {model} ::')
    print(f'r2_score is {r2_score(y,y_predict_train)}')
    print(f'Mean squared error {mean_squared_error(y,y_predict_train)}')
    print(f'Root mean square error {np.sqrt(mean_squared_error(y,y_predict_train))}')
    print(f'Mean absolute error {mean_absolute_error(y,y_predict_train)}')
    print(model.get_params())
    print('\n')
    
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor

models = [LinearRegression(),Ridge(),Lasso(),SVR(),RandomForestRegressor()]
```

```python
for m in models:
    model_training(m,x_train, y_train)
    
```

 ## Using RandomForest Regression with default parameters on test Data to check for overfitting

```python
random_forest = RandomForestRegressor()
random_forest.fit(x_train,y_train)

y_test_predict = random_forest.predict(x_test)

print('For test data using Random Forest')
print(f'r2_score is {r2_score(y_test,y_test_predict)}')
print(f'Mean squared error {mean_squared_error(y_test,y_test_predict)}')
print(f'Root mean square error {np.sqrt(mean_squared_error(y_test,y_test_predict))}')
print(f'Mean absolute error {mean_absolute_error(y_test,y_test_predict)}')
```

## Above results show that the model is slighlty overfitting, hence we need to tweak the parameters

## Using GridSearchcv to tweak paramters of RandomForest

```python
parameters = {'n_estimators': [50,100,150,200,250,300,350,400,450,500,550],
              'max_depth': ['None',4,5,6,7,8,9,10],
              'bootstrap' :[True],
              'oob_score':[True,False], 
              'n_jobs' :[-1]}
```

```python
from sklearn.model_selection import GridSearchCV
random_forest = RandomForestRegressor()
randomforrest_grid_search = GridSearchCV(estimator=random_forest,
                                        param_grid=parameters,
                                        scoring='neg_mean_squared_error',
                                        cv = 2,verbose = 2
                                        )
```

```python
randomforrest_grid_search.fit(x_train,y_train)
```

```python
randomforrest_grid_search.best_estimator_
```

```python
y_train_pred_grid_search = randomforrest_grid_search.predict(x_train)
```

```python
print('For training data using GridSearchCV using Random Forest')
print(f'r2_score is {r2_score(y_train,y_train_pred_grid_search)}')
print(f'Mean squared error {mean_squared_error(y_train,y_train_pred_grid_search)}')
print(f'Root mean square error {np.sqrt(mean_squared_error(y_train,y_train_pred_grid_search))}')
print(f'Mean absolute error {mean_absolute_error(y_train,y_train_pred_grid_search)}')
```

```python
y_test_pred_grid_search = randomforrest_grid_search.predict(x_test)
```

```python
print('For test data using GridSearchCV using Random Forest')
print(f'r2_score is {r2_score(y_test,y_test_pred_grid_search)}')
print(f'Mean squared error {mean_squared_error(y_test,y_test_pred_grid_search)}')
print(f'Root mean square error {np.sqrt(mean_squared_error(y_test,y_test_pred_grid_search))}')
print(f'Mean absolute error {mean_absolute_error(y_test,y_test_pred_grid_search)}')
```
