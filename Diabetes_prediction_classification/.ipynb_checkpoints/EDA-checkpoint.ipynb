```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('diabetes.csv')
dataset.head()
```

## About the Dataset ::

We have a dataset which classifies whether a patient will have diabetes or not based on some input features such as Pregnancies,Glucose
BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction and age .

Based on mentioned featured we will predict whether a patient has diabetes or not .
1 indicates presence of Diabetes and 0 indicates absence of diabetes



```python
dataset.isnull().sum() #### checking for null values in any column 
```

```python
## Plotting the data distribution of each independent feature  except the dependent feature
for cols in dataset.columns[:-1]:
    sns.histplot(dataset[cols],kde=True)
    plt.show()
```

```python
## Checking whether the Dataset is balanced or imbalanced 
sns.countplot(dataset['Outcome'])
plt.show()
```

```python
### Checking for outliers through boxplot
for i in dataset.columns[:-1]:
    sns.boxplot(dataset[i])
    plt.show()
```

```python
dataset.columns
    
```

```python
for j in dataset.columns[:-1]:
    for k in dataset.columns[:-1]:
        sns.scatterplot(data=dataset, x=j, y=k , hue='Outcome')
        plt.show()
```

```python
## Insights from the Dataset ::

1. Dependent feature is 'Outcome' which needs to be predicted
2. Independent features are Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction and age
3. Majority of the data is from the age 20-45
4. People having high glucose, high blood pressure & high BMI tend to have diabetes in more number .
5  Peeple having high glucose level tend to have diabetes in more number.
6. There is no missing values
7. We are dealing with an imbalanced dataset
```
